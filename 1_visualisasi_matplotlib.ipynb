```python
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y)
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.title('judul')
plt.show()
```

```python
plt.subplot(121)
plt.plot(x,y,color='red')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')

plt.subplot(122)
plt.plot(y,x, color='blue')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.tight_layout() 

plt.show()
```
