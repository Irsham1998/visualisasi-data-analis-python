```python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set()
```

```python
dataku = pd.read_csv('train.csv')
```

```python
dataku.head()
```

```python
#memeriksa data string
from xml.etree.ElementInclude import include


dataku.describe(include='O')
```

```python
#barplot (histogram)
sns.barplot(x='Sex', y='Age', data=dataku);
```

```python
sns.barplot(x='Fare', y='Sex', data=dataku);
```

```python
sns.barplot(x='Sex', y='Fare', data=dataku, estimator=sum);
```

```python
sns.barplot(x='Sex', y='Fare', data=dataku, estimator=np.std); #standar deviasi
```

```python
#menghitung data | countplot
sns.countplot(x='Sex', data=dataku);
```

```python
sns.countplot(x='Embarked', hue='Sex', data=dataku);
```

```python
sns.countplot(x='Embarked', hue='Sex', order=['C','S','Q'], data=dataku); #diatur munculya berurut
```

```python
sns.boxplot(x='Sex', y='Age', hue='Embarked', data=dataku); #hue_order=[urutkan]
```

```python
#violin plot
sns.violinplot(x='Sex', y='Age', data=dataku);
```

```python
#stripplot
sns.stripplot(x='Embarked', y='Age', hue='Sex', jitter=True, data=dataku);
```

```python
#swarmplot
sns.swarmplot(x='Embarked', y='Age', hue='Sex', data=dataku);
```
