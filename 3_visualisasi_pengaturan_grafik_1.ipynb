```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,20,0.5)
y = x**2
```

**Mengatur Figsize dan DPI**

```python
#versi 1
fig =plt.figure(figsize=(8,4), dpi=(100)) #ukuran dalam inch || 300 dpi paling ok
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

```python
#versi 2
fig, axes = plt.subplots(2,1, figsize=(4,4), dpi=(100))

axes[0].plot(x,y)
axes[0].set_title('Grafik 1')

axes[1].plot(y,x)
axes[1].set_title('Grafik 2')

plt.tight_layout()

plt.show()
```

```python
#ketika ingin dicetak
fig.savefig('tes2.png', dpi=300)
```
