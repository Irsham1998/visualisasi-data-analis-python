```python
import pandas as pd
import seaborn as sns
sns.set()
```

```python
dataku = pd.read_csv('train.csv')
```

```python
dataku.head()
```

```python
#membuat heatmap = matriks plot hanya bisa menggunakan data numerik
```

```python
#membuat tabel korelasi
korelasi = dataku.corr()
korelasi
```

```python
#membuat heatmap
sns.heatmap(korelasi);
```

```python
sns.heatmap(korelasi, annot=True); #menambahkan keterangan anotasi
```

```python
sns.heatmap(korelasi, annot=True, cmap='Reds'); #mengganti warna
```

```python
terbang = sns.load_dataset('flights') #import dataset bawaan seaborn
```

```python
terbang
```

```python
#membuat pivot tabel
terbang_pivot = terbang.pivot_table(index='month', columns='year', values='passengers')
```

```python
terbang_pivot
```

```python
sns.heatmap(terbang_pivot, cmap='Reds');
```

```python
sns.heatmap(terbang_pivot, cmap='Reds', linecolor='black', linewidths=1); #menambah variasi
```

```python
#proses clustering map
sns.clustermap(terbang_pivot)
```

```python
sns.clustermap(terbang_pivot, cmap='coolwarm')
```

```python
sns.clustermap(terbang_pivot, cmap='coolwarm', standard_scale=1)
```
