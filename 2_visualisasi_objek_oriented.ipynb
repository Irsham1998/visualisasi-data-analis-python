```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plt.plot() versi konvensional
```

```python
x = np.arange(1,20,0.5)
y = x**2
```

```python
#tampilkan diagram versi OOP
#versi 1 oop
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #jarak kiri, jarak bawah,lebar,tinggi
axes.plot(x,y)
axes.set_xlabel('sumbu X')
axes.set_ylabel('sumbu Y')
axes.set_title('Judul Diagram')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y,color='red')
axes1.set_xlabel('X besar')
axes1.set_ylabel('Y besar')
axes1.set_title('Judul Utama')

axes2.plot(y,x,color='blue')
axes2.set_xlabel('X kecil')
axes2.set_ylabel('Y kecil')

plt.show()
```

```python
#versi 2 OOP
```

```python
fig, axes = plt.subplots(1,2) #baris dan kolom

#plot 1
axes[0].plot(x,y)
axes[0].set_xlabel('Sumbu X1')
axes[0].set_ylabel('Sumbu Y1')
axes[0].set_title('Judul 1')

#plot 2
axes[1].plot(y,x)
axes[1].set_xlabel('Sumbu X2')
axes[1].set_ylabel('Sumbu Y2')
axes[1].set_title('Judul 2')

fig.tight_layout()
```
