```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,20,0.5)
y = x**2
```

**Menambahkan Legenda**

```python
fig = plt.figure(facecolor='orange',dpi=100)
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='pangkat 2')
ax.plot(x,x**3, label='Pangkat 3')

ax.legend(loc=('upper center')) #0.5 artinya 50% = legend(loc=(0.5,0.5))
ax.grid(True) #baca dokumentasi di matplotlib

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='pangkat 2', lw=1, ls='-') #linewidth/lw baca dokumentasi bisa di singkat
ax.plot(x,x**3, label='Pangkat 3', lw=2, ls='--')
ax.plot(x,x**4, label='Pangkat 4', lw=3, ls='-.')

ax.legend(loc=('upper left')) #0.5 artinya 50% = legend(loc=(0.5,0.5))
ax.grid(True) #baca dokumentasi di matplotlib

plt.show()
```

```python
fig = plt.figure(dpi=300)
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='pangkat 2', lw=1, ls='-', marker='o', markersize=5, mfc='yellow', mew=2, mec='red') 
ax.plot(x,x**3, label='Pangkat 3', lw=2, ls='--', marker='*', markersize=5, mfc='yellow', mew=2, mec='red')
ax.plot(x,x**4, label='Pangkat 4', lw=3, ls='-.', marker='+', markersize=5, mfc='yellow', mew=2, mec='red')

ax.legend(loc=(0.5,0.5))
ax.grid(True) #baca dokumentasi di matplotlib

plt.show()
```

```python
fig = plt.figure(dpi=100)
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='pangkat 2', lw=1, ls='-', marker='o', markersize=5, mfc='yellow', mew=2, mec='red') 
ax.plot(x,x**3, label='Pangkat 3', lw=2, ls='--', marker='*', markersize=5, mfc='yellow', mew=2, mec='red')
ax.plot(x,x**4, label='Pangkat 4', lw=3, ls='-.', marker='+', markersize=5, mfc='yellow', mew=2, mec='red')

ax.legend(loc=(0.5,0.5))
ax.grid(True) #baca dokumentasi di matplotlib

ax.set_xlim([0,12.5])
ax.set_ylim([0,20000])

plt.show()
```
