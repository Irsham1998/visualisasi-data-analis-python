```python
sns.PairGrid(berlian)
```

```python
#sebut saja P
p = sns.PairGrid(berlian)
p.map(plt.scatter)
plt.show()
```

```python
p = sns.PairGrid(berlian) #custom grid
p.map_diag(sns.histplot) #mengatur diagonal histplot atau displot
p.map_upper(plt.scatter) #mengatur agar bagian atas isinya scatter
p.map_lower(sns.kdeplot) #mengatur bagian bwah isinya kdeplot
```

FacetGrid

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
t = sns.FacetGrid(tips, row='smoker', col='day')
t.map(sns.histplot, 'total_bill')
plt.show()
```

```python
t = sns.FacetGrid(tips, row='smoker', col='time')
t.map(sns.histplot, 'total_bill', kde=True)
plt.show()
```

```python
t = sns.FacetGrid(tips, row='smoker', col='time')
t.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```
